### Bài tập 1:
**Câu hỏi**: Hãy viết thuật toán chọn lớp trưởng trong lớp học theo các bước sau:

    1. Lập danh sách các sinh viên
    2. Sắp xếp danh sách
    3. Chọn người đứng đầu danh sách làm lớp trưởng

<center>
Bài làm
</center>

- Bước 1. Lập danh sách m sinh viên có i thuộc tính (ví dụ như là gpa, drl, ... tương ứng cho thuộc tính thứ 1, 2, ..., i) và đặt n = 1
- Bước 2. Sắp xếp danh sách theo thứ tự cao đến thấp theo thuộc tính thứ n
- Bước 3. Chọn người đứng đầu danh sách làm lớp trưởng tạm thời (tương ứng j = 1)
- Bước 4. Xét thuộc tính thứ n của người đứng đầu danh sách (tương ứng j = 1) với từng người còn lại trong danh sách (tương ứng k = 2 đến m) từ trên xuống dưới
    - Nếu thuộc tính thứ n của người đứng đầu danh sách > thuộc tính thứ n của người thứ k thì bỏ người thứ k trở xuống khỏi danh sách và đặt m = k - 1
    - Nếu toàn bộ người còn lại bằng thì qua bước 5
- Bước 5. Nếu m = 1 thì dừng lại, người đứng đầu danh sách là lớp trưởng, kết thúc bài toán
    - Nếu m > 1 thì
        - Nếu n < i thì đặt n = n + 1 (tức là khi này xét thuộc tính tiếp theo) và quay lại bước 2
        - Nếu n = i thì dừng lại, người đứng đầu danh sách là lớp trưởng, kết thúc bài toán

### Bài tập 2:
**Câu hỏi**: 
  - Input: a,b,c
 - Giải phương trình bậc 2 ax^2 + bx + c = 0

##

<center>
Bài làm
</center>

- Bước 1. Nhập a, b, c
- Bước 2. Kiểm tra nếu a = 0 thì chuyển sang giải phương trình bậc nhất bx + c = 0
    - Nếu b = 0 thì
        - Nếu c = 0 thì phương trình có vô số nghiệm, kết thúc bài toán.
        - Nếu c != 0 thì phương trình vô nghiệm, kết thúc bài toán.
    - Nếu b != 0 thì phương trình có một nghiệm duy nhất x = -c/b, kết thúc bài toán.
- Bước 3. Tính delta = b^2 - 4ac
- Bước 4. Kiểm tra delta
    - Nếu delta < 0 thì phương trình vô nghiệm, kết thúc bài toán.
    - Nếu delta = 0 thì phương trình có nghiệm kép x = -b/(2a), kết thúc bài toán.
    - Nếu delta > 0 thì phương trình có hai nghiệm phân biệt x1 = (-b + sqrt(delta))/(2a) và x2 = (-b - sqrt(delta))/(2a), kết thúc bài toán.
