Phân tích thuật toán sau:
```
i = 1;
res = 0;
s = 0; // so thuc
while i <= n do
    j = 1;
    s = s + 1/i;
    while j <= s do
        res = res + i*j;
        j = j + 1;
    endw
    i = i + 1;
endw
```


Trước hết ta có 3 phép gán:
```
i = 1;
res = 0;
s = 0;
```
Trong vòng lặp while i <= n thì mỗi vòng lặp ta luôn luôn có 3 phép gán sau:
```
j = 1;
s = s + 1/i;
i = i + 1;
```
Vì vậy, ta có 3n phép gán trong vòng lặp while i <= n.

Nhưng mà do s = s + 1/i nên s sẽ tăng dần theo i nên ở vòng lặp tiếp theo while j <= s trong vòng lặp while i <= n sẽ có số lần lặp là s = s + 1/i, tức là s sẽ tăng dần theo i. Do đó, số lần lặp của vòng while j <= s sẽ tăng dần theo i.
Tức là $$s = 1 + 1/2 + 1/3 + ... + 1/i = \sum_{k=1}^{i} 1/k$$

Mà trong vòng lặp while j <= s thì có 2 phép gán là: 
```
res = res + i*j;
j = j + 1;
```
Do đó, số phép gán trong vòng lặp while j <= s sẽ là $$2\sum_{k=1}^{i} 1/k$$
Khi kết hợp với vòng lặp while i <= n thì số phép gán trong vòng lặp while j <= s sẽ là $$2\sum_{i=1}^{n} \sum_{k=1}^{i} 1/k$$
Như vậy tổng số phép gán trong thuật toán sẽ là:
$$3+ 3n + 2\sum_{i=1}^{n} \sum_{k=1}^{i} 1/k$$
