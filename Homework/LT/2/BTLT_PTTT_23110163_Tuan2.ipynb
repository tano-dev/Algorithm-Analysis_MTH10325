### Bài tập: 


**Câu hỏi**: Hãy chứng minh với mọi y, multiply(y, z) = y*z với mọi z lớn hơn 0.

<center>
Bài làm
</center>

- Xét khi $z = 0$ thì while z > 0 không chạy và trả về $x = 0$ nên multiply(y, 0) = 0 = y * 0 đúng với mọi y.
- Xét khi $z = 1$ thì while chạy 1 lần: $1$ là số lẻ nên $x = 0 + y = y$ rồi $y = 2y$ và $z = [1/2] = 0$. Tiếp theo while kết thúc (do z = 0) và trả về $x = y$. Vậy multiply(y, 1) = y = y * 1 đúng với mọi y

- Ta giả sử với mọi $y$, multiply(y, k) = y * k đúng với mọi $k$ sao cho $0 \le k < n$ (với $n > 1$).
    - Xét số $n > 1$. Ta có thể biểu diễn $n = 2m + r$ (với $r \in \{0, 1\}$ và $m = [n / 2]$, m < n). Khi $z = n$ thì ta xét 2 trường hợp của $r$:
        - Nếu $r = 0$ (tức là $n$ chẵn, $n = 2m$): If không chạy, $x$ vẫn bằng $0$.Khi đó cập nhật $y= 2y$ và $z= m$. Ta thấy while từ thời điểm này trở đi hoàn toàn tương đương với multiply(2y, m). Kết hợp với giả sử ở trên (vì $m < n$), ta có:
        $$\text{multiply}(2y, m) = (2y) \cdot m = y \cdot (2m) = y \cdot n$$
        - Nếu $r = 1$ (tức là $n$ lẻ, $n = 2m + 1$): If chạy, $x = x + y$ (lúc này $x = y$).Khi đó cập nhật $y= 2y$ và $z= m. Ta thấy while từ thời điểm này trở đi hoàn toàn tương đương với multiply(2y, m). While chạy cho tới khi z = 0 thì khi này ta có kết quả x chính là y * n. Kết hợp với giả sử ở trên, ta có:
        $$x + \text{multiply}(2y, m) = y + (2y) \cdot m = y + y \cdot (2m) = y \cdot (2m + 1) = y \cdot n$$

Vậy với mọi $y$, multiply(y, n) = y * n đúng với mọi $n \ge 0$.
